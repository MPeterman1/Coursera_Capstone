### Part 1 - Web Scraping off Wikipedia with BeautifulSoup

```python
# import the library we use to open URLs
import urllib.request
# grab page
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
page = urllib.request.urlopen(url)
```

```python
# import the BeautifulSoup library so we can parse HTML and XML documents
from bs4 import BeautifulSoup
# parse the HTML from our URL into the BeautifulSoup parse tree format
soup = BeautifulSoup(page, "lxml")
print(soup.prettify())
```

```python
soup.title.string
```

```python
#Tables listed on the web page
all_tables=soup.find_all("table")
all_tables
```

```python
right_table=soup.find('table', class_='wikitable sortable')
right_table
```

```python
#Loop through table to make lists of column data
A=[]
B=[]
C=[]

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

### Part 2 - Create and Clean Pandas Dataframe

```python
import pandas as pd
df=pd.DataFrame(A,columns=['Postcode'])
df['Borough']=B
df['Neighborhood']=C
df
```

```python
# Remove Boroughs that are not assigned.
indexNames = df[df['Borough'] == "Not assigned" ].index
 
# Delete these unassigned row indexes from dataFrame
df.drop(indexNames , inplace=True)
df
```

```python
df=df.groupby(['Postcode','Borough'])['Neighborhood'].apply(', '.join).reset_index()
df
```

```python
df.shape
```

### Part 3 - Add Latitude and Longitude to Dataframe

```python
df_geo=pd.read_csv("https://cocl.us/Geospatial_data")
df_geo
#Key column names do not match. Rename column names before joining to avoid duplicate columns
df_geo=df_geo.rename(columns={"Postal Code": "Postcode"})
df_geo
```

```python
#Join the dataframes to get result
result=df.merge(df_geo[['Postcode','Latitude', 'Longitude']], left_on = ['Postcode'], right_on = ['Postcode'], how = 'outer').fillna('')
result

# Remove Boroughs that are not assigned.
indexNames = result[result['Borough'] == "Not assigned"].index
 # Rename Neighbourhood to Neighborhood
result=result.rename(columns={"Neighbourhood": "Neighborhood"})
# Delete these unassigned row indexes from dataFrame
result.drop(indexNames , inplace=True)
result
```

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")
toronto_onehot.drop(['Neighborhood'],axis=1,inplace=True) 
toronto_onehot.insert(loc=0, column='Neighborhood', value=toronto_venues['Neighborhood'] )

toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped.head()
```

```python
## Get the top 10 venues in the Boroughs named with "Toronto" in the name
### Sort the venues in descending order
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 5

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

## Cluster the neighborhooods

```python
toronto_merged=result

# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe

labels = kmeans.labels_
print(labels)
```

```python
toronto_merged = result

# add clustering labels
toronto_merged['Cluster Labels'] = labels

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = result.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

toronto_merged.head()
```

```python
toronto_merged=toronto_merged.dropna()
```

## MAP and Explore Top 5 Venue Clusters

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=12)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.2).add_to(map_clusters)
       
map_clusters
```
